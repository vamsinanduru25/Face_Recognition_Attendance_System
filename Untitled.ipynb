```python
import os
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.2, random_state=42)

# Initialize empty lists for accuracy and neighbors
accuracy_list = []
neighbors_list = []

# Try different numbers of neighbors and calculate accuracy
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    accuracy_list.append(accuracy)
    neighbors_list.append(k)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot accuracy as a function of number of neighbors
ax.plot(neighbors_list, accuracy_list)

# Add labels and title
ax.set_xlabel('Number of Neighbors')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of KNN Model')

# Display the plot
plt.show()
```

```python
import os
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, classification_report

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.2, random_state=42)

# Train a KNN model with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = knn.predict(X_test)

# Calculate confusion matrix, F1 score, and classification report
conf_mat = confusion_matrix(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='macro')
class_report = classification_report(y_test, y_pred)

# Print the results
print('Confusion matrix:\n', conf_mat)


```

```python
import os
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from prettytable import PrettyTable

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.2, random_state=42)

# Fit the KNN model on the training set
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluate the accuracy of the KNN model on the testing set
accuracy = knn.score(X_test, y_test)
print(f'Accuracy: {accuracy}')

# Generate classification report
y_pred = knn.predict(X_test)
class_report = classification_report(y_test, y_pred, output_dict=True)

# Generate table for classification report
table = PrettyTable()
table.field_names = ["Class", "Precision", "Recall", "F1-score", "Support"]
for cls, metrics in class_report.items():
    if cls not in ['accuracy', 'macro avg', 'weighted avg']:
        table.add_row([cls, metrics['precision'], metrics['recall'], metrics['f1-score'], metrics['support']])

# Print the table
print(table)
```

```python
import os
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tabulate import tabulate

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.2, random_state=42)

# Fit the KNN model on the training set
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluate the accuracy of the KNN model on the testing set
accuracy = knn.score(X_test, y_test)
print(f'Accuracy: {accuracy}')

# Generate classification report
y_pred = knn.predict(X_test)
class_report = classification_report(y_test, y_pred, output_dict=True)

# Generate table for classification report
table = []
table.append(['Class', 'Precision', 'Recall', 'F1-score', 'Support'])
for cls, metrics in class_report.items():
    if cls not in ['accuracy', 'macro avg', 'weighted avg']:
        table.append([cls, metrics['precision'], metrics['recall'], metrics['f1-score'], metrics['support']])

# Print the table
print(tabulate(table, headers='firstrow'))
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import time

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.2, random_state=42)

# Fit the KNN model on the training set and record the time taken
start_time = time.time()
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
end_time = time.time()
training_speed = end_time - start_time

# Evaluate the accuracy of the KNN model on the testing set
y_pred = knn.predict(X_test)
accuracy = knn.score(X_test, y_test)
print(f'Accuracy: {accuracy}')

# Generate classification report and convert to pandas DataFrame
class_report = classification_report(y_test, y_pred, output_dict=True)
df = pd.DataFrame(class_report).transpose()

# Print the training speed and classification report DataFrame
print(f'Training speed: {training_speed} seconds')
print(df)
```

```python
import os
import cv2
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# Initialize empty lists for faces and labels
faces = []
labels = []

# Load face images and corresponding labels
userlist = os.listdir('static/faces')
for user in userlist:
    for imgname in os.listdir(f'static/faces/{user}'):
        img = cv2.imread(f'static/faces/{user}/{imgname}')
        resized_face = cv2.resize(img, (50, 50))
        faces.append(resized_face.ravel())
        labels.append(user)

# Convert faces and labels to NumPy arrays
faces = np.array(faces)
labels = np.array(labels)

# Set the number of training samples
num_samples = [100, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000]

# Initialize empty list for training times
training_times = []

# Loop over the number of training samples
for num in num_samples:
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(faces[:num], labels[:num], test_size=0.2, random_state=42)
    
    # Fit the KNN model on the training set and calculate training time
    start_time = time.time()
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)
    end_time = time.time()
    
    # Append the training time to the list
    training_times.append(end_time - start_time)

# Plot a graph of training time vs number of training samples
plt.plot(num_samples, training_times)
plt.xlabel('Number of training samples')
plt.ylabel('Training time (seconds)')
plt.show()
```
